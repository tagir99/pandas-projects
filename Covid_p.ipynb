```python
import numpy as np

st=np.array(df['Краткие обозначения'])
print(len(st))
df_t = np.array(df['Наборы тестов (кол-во)'].fillna(0))
print(len(df_t))
```

```python
ind_z = []
for var in range(len(df_t)):
    if df_t[var]==0:
        ind_z.append(var)
ind_z
```

```python
# Создать список с индексами значений, которые равны 0
# удалить значения с этим индексом из списка с обозначеиями и списка - ['Наборы тестов (кол-во)']

st=np.array(df['Краткие обозначения'])
print(len(st))
df_t = np.array(df['Наборы тестов (кол-во)'].fillna(0))
print(len(df_t))

ind_z = []
for var in range(len(df_t)):
    if df_t[var]==0:
        ind_z.append(var)
ind_z

st = np.delete(st, ind_z)
print(st)
print()
print(df_t)
df_t = np.delete(df_t, ind_z)
print(df_t)

```

```python
plt.figure(figsize=(10,5))
plt.bar(st, df_t)
plt.show()
```
